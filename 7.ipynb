```python
#W7A3
c = list(map(int, input().split()))
def bubble(c):
  sorted = False
  indexfin = len(c) - 1

  while not sorted:
    sorted = True
    for i in range(0, indexfin):
      if c[i] > c[i + 1]:
        sorted = False
        c[i], c[i+1] = c[i+1], c[i]

  return c

print(bubble(c))

def selection(c):
  for i in range(0, len(c) - 1):
    min_val = i
    for j in range(i+1, len(c)-1):
      if c[j] < c[min_val]:
        min_val = j

    if min_val != i:
      c[i], c[min_val] = c[min_val], c[i]

  return c

print(selection(c))
```

```python
#W7A4
c = list(map(int, input().split()))
d_test = {}
for i in range(len(c)):
  if c[i] not in d_test:
    d_test[c[i]] = 1
  else:
    d_test[c[i]] += 1

max = 0
for j in d_test.values():
  if j > max:
    max = j

max_k = 0
for a in d_test.keys():
  if d_test[a] == max:
    max_k = a
    break

print(f"{max_k} xuat hien nhieu nhat, som nhat, {max} lan")
```

```python
#W7A6
import ast

data = ast.literal_eval(input())
a_flat = []

for item in data:
    if isinstance(item, list):
        a_flat.extend(item)
    else:
        a_flat.append(item)

print(a_flat)
```

```python
#W7A7
import ast
list_numb = ast.literal_eval(input())
remember = []

for i in range(0, len(list_numb)):
  count = 0
  for j in range(i+1, len(list_numb)):
    if list_numb[i] < list_numb[j]:
      count += 1
      i = j
  remember.append(count)

print(max(remember)+1)
```
