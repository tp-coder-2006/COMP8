```python
#W5A3
a = int(input("Nhập số nguyên a:"))
def prime(a):
  for i in range(2, int(a**0.5)+1):
    if a%i==0:
      return False
  return True

prime(a)
```

```python
#W5A4
a = int(input("Nhập số nguyên a:"))
def perfect(a):
  sum = 0
  for i in range(1, a):
    if a%i==0:
      sum += i
  if sum == a:
    return True
  else:
    return False

perfect(a)
```

```python
#W5A5
def firstnumb(input_list):
  if not input_list:
    return -1
  fin = input_list[-1]
  for i in range(len(input_list) - 1):
    if fin == input_list[i]:
      return i+1
  return -1

a = __builtins__.list(map(int, input().split()))
print(firstnumb(a))
```

```python
#W5A6
def giaithua(a):
  tich = 1
  for i in range(1, a+1):
    tich*=i
  return tich

a = int(input("Nhập số nguyên a:"))
giaithua(a)
```

```python
#W5A7
def tinhtoan(a):
  a[0] = int(a[0])
  a[2] = int(a[2])
  b = 0
  if a[1] == '/':
    b = a[0] / a[2]
  elif a[1] == '*':
    b = a[0] * a[2]
  elif a[1] == '+':
    b = a[0] + a[2]
  elif a[1] == '-':
    b = a[0] - a[2]
  return round(b, 2)

a = list(map(str, input().split()))
tinhtoan(a)
```

```python
#W5A10
def dangcau(a, b):
  A = a.split()
  B = b.split()

  if len(A) != len(B):
    return False

  for i in range(len(A)):
    c = A[i]
    for j in range(i, len(A)):
      if c == A[j]:
        if B[j] == B[i]:
          continue
        else:
          return False
  return True

a, b = map(str, input().split())
print(dangcau(a, b))
```
