```python
#W6A1
def rev_same_correct(a):
    seen = set()
    result_list = []

    for element in a:
        if element not in seen:
            result_list.append(element)
            seen.add(element)

    return result_list

a = list(map(int, input().split()))
print(rev_same_correct(a))
```

```python
#W6A2
'''
Nhập một dòng số nguyên. In ra list tổng tích lũy
(phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ
tự
'''
def tongtichluy(a):
  new = []
  for i in range(0, len(a)):
    new.append(sum(a[0:(i+1)]))
  return new

a = list(map(int, input().split()))
print(tongtichluy(a))
```

```python
#W6A3
'''
Nhập một tuple số nguyên (một dòng) và số nguyên
k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ
nguyên thứ tự tương đối sau xoay; không sắp xếp.)
'''
def xoayk(a, k):
  newt = ()
  if len(a) == 1:
    return a
  for i in range(k, len(a)):
    test = (a[i], )
    newt += test
  if k == 0:
    return newt
  for i in range(0, k):
    test = (a[i], )
    newt += test
  return newt

a = tuple(map(int, input().split()))
k = int(input())
print(xoayk(a, k))
```

```python
#W6A4
'''
Nhập một dòng gồm các cặp key:value cách nhau
bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list
các value theo thứ tự xuất hiện. Nếu key lặp lại,
thêm vào list của key đó
'''
def tao_dict_cua_lists(data_string):
    result_dict = {}

    cap_key_value = data_string.split()

    for cap in cap_key_value:
        parts = cap.split(":", 1)

        if len(parts) == 2:
            key = parts[0].strip()
            value = parts[1].strip()
            result_dict.setdefault(key, []).append(value)

    return result_dict

data_input = input()
ket_qua = tao_dict_cua_lists(data_input)
print(ket_qua)
```

```python
#W6A4
'''
Nhập một dòng gồm các cặp key:value cách nhau
bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list
các value theo thứ tự xuất hiện. Nếu key lặp lại,
thêm vào list của key đó
'''
a = list(input().split(" "))
d = {}
for i in a:
  b, c = i.split(":")
  if b not in d:
    d[b] = [c]
  elif b in d:
    d[b].append(c)

print(d)
```

```python
#W6A5
'''
Nhập một list số nguyên, In ra dictionary {'positives':
..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)
'''
pnz = {}
n_i = list(map(int, input().split()))
pnz["positives"] = 0
pnz["negatives"] = 0
pnz["zeros"] = 0
for i in n_i:
  if i > 0:
    pnz["positives"] += 1
  elif i < 0:
    pnz["negatives"] += 1
  else:
    pnz["zeros"] += 1

print(pnz)
```

```python
#W6A10
'''
Viết chương trình nhập vào một danh sách các số nguyên,
và số nguyên dương k. In ra list gồm các cặp số khác
nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong
tuple có thể giống nhau, thứ tự xuất hiện trong danh sách
của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple
trong list được sắp xếp theo thứ tự tăng dần của phần tử
đầu tiên trong tuple.
'''
n = list(map(int, input().split()))
k = int(input())
outp = []
for i in range(0, len(n)):
  for j in range(i+1, len(n)):
    if j > (len(n)-1):
      break
    if n[i] + n[j] == k:
      if (n[i], n[j]) not in outp:
        outp.append((n[i], n[j]))
print(sorted(outp))
```

```python
#W6A13
'''
[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất
(không trùng nhau). Hãy tạo một dictionary mới
bằng cách đảo ngược key và value của dictionary
ban đầu. In ra dictionary mới.
'''
a = list(input().split())
d = {}
for i in range(0, len(a)):
  if i % 2 == 0:
    d[a[i]] = None
  elif i % 2 == 1:
    d[a[i-1]] = a[i]
e = {}
for i, j in d.items():
  e[j] = i
print(e)
```

```python
#W6A14
'''
[Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các
số nguyên. Hãy tìm các phần tử chung (giao) của
hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp.
'''
a = list(map(int, input().split()))
b = list(map(int, input().split()))

ce = list(set(a).intersection(set(b)))

print(ce)
```

```python
#W6A15
'''
[Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là
số nguyên, và một số nguyên k ở dòng tiếp theo.
Hãy tạo một dictionary mới chỉ chứa các cặp key-value từ dictionary ban đầu mà có value lớn hơn k.
In ra dictionary mới.
'''
a = list(input().split())
k = int(input())
b = {}
for i in range(0, len(a)):
  if i % 2 == 0:
    b[a[i]] = 0
  elif i % 2 == 1:
    b[a[i-1]] = int(a[i])
c = {}
for x, y in b.items():
  if y > k:
    c[x] = y
print(c)



```
